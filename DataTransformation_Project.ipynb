```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
car_df = pd.read_csv('../data/cars.csv')
```

```python
brand_df = pd.read_csv('../data/brand.csv')
```

```python
car_df.head()
```

```python
brand_df.head()
```

```python
car_df['title'].unique()
```

```python
brand_df['title'].unique()
```

```python
# cars.csv와 brand.csv의 데이터 합치기
# car_df의 title과 brand_df의 title 같게 만들기
car_df = car_df.assign(title_brand=car_df['title'].str.lower().str.split().str[0])
```

```python
car_df.info()
```

```python
#car_df와 brand_df 합치기
CarWithBrand_df=car_df.merge(brand_df,left_on = 'title_brand',right_on='title', how='outer')
```

```python
CarWithBrand_df.head()
```

```python
#중복컬럼 제거
CarWithBrand_df.rename({'title_x':'title'},axis=1,inplace=True)
```

```python
CarWithBrand_df.head()
```

```python
CarWithBrand_df[CarWithBrand_df['Previous Owners'].isna()]
#중고차 데이터인데 이전 주인이 왜 null값인지? 
```

```python
#카테고리형 변수 숫자로 변경하기
#Engine
CarWithBrand_df['Engine'] = CarWithBrand_df['Engine'].astype(str).str.replace('L', '', regex=False).str.strip()
CarWithBrand_df['Engine'] = pd.to_numeric(CarWithBrand_df['Engine'], errors='coerce')
```

```python
#Emission Class
CarWithBrand_df['Emission Class'] = CarWithBrand_df['Emission Class'].astype(str).str.replace('Euro', '').str.strip()
CarWithBrand_df['Emission Class'] = pd.to_numeric(car_df['Emission Class'],errors='coerce')
```

```python
CarWithBrand_df.info()
```

```python
#결측치 처리하기
# NaN값 존재 = Previous Owners, Engine, Doors, Seats, Emission Class, Service history
# 이 중 Engine,Doors,Seats,Emission Class만 dropna() 하기로 결정
# Previous Owners가 NaN인 것의 비율은 약 0.4로 너무 많아서 날릴 수 없음 -> Unknown로 채우기
# Service history는 Unknown로 채우기
 
```

```python
CarWithBrand_df['Previous Owners'].isna().mean()
```

```python
CarWithBrand_df['Emission Class'].isna().mean()
```

```python
CarWithBrand_df = CarWithBrand_df.dropna(subset=['Seats'])
```

```python
CarWithBrand_df = CarWithBrand_df.dropna(subset=['Doors'])
```

```python
CarWithBrand_df = CarWithBrand_df.dropna(subset=['Engine'])
```

```python
CarWithBrand_df = CarWithBrand_df.dropna(subset=['Emission Class'])
```

```python
# Previous Owners는 Unknown로 채우기
CarWithBrand_df['Previous Owners'] = CarWithBrand_df['Previous Owners'].fillna('Unknown')
```

```python
# Service history는 Unknown로 채우기
CarWithBrand_df['Service history'] = CarWithBrand_df['Service history'].fillna('Unknown')
```
