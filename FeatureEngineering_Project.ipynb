```python
# Frequency Encoding(빈도 기반) by ChatGPT
# 가맹점별 거래횟수를 count한 새로운 피쳐 생성
merchant_counts = cc_new_df['merchant'].value_counts()
cc_new_df['merchant_freq'] = cc_new_df['merchant'].map(merchant_counts)
```

거래 시간 차이, 거래 위치 차이 계산

```python
# 1. 거래 시간 차이 계산

# `cc_num`별 거래를 시간 순으로 정렬
cc_new_df = cc_new_df.sort_values(by=['cc_num', 'trans_date_trans_time'])

# 거래 시간 차이 (초 단위) 계산
cc_new_df['time_diff'] = cc_new_df.groupby('cc_num')['trans_date_trans_time'].diff().dt.total_seconds()

#초(second) 단위 차이를 시:분:초 형식으로 변환
cc_new_df['time_diff_hms'] = pd.to_timedelta(cc_new_df['time_diff'], unit='s')
```

```python
# 2. 거래 위치 차이 계산

# 위도 및 경도 차이 계산
cc_new_df['lat_diff'] = cc_new_df.groupby('cc_num')['merch_lat'].diff()
cc_new_df['long_diff'] = cc_new_df.groupby('cc_num')['merch_long'].diff()

# 유클리드 거리 계산
cc_new_df['distance_diff'] = np.sqrt(cc_new_df['lat_diff']**2 + cc_new_df['long_diff']**2)
```
